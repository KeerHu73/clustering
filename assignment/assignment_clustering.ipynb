3. Create a scree plot for each of the datasets. Describe how the level of `noise` affects the scree plot (particularly the presence of a clear "elbow") and your ability to definitively select a $k$. (Pay attention to the vertical axis across plots, or put all the scree curves on a single canvas.)

````python
import numpy as np
import pandas as pd

def createData(noise,N=50):
    np.random.seed(100) # Set the seed for replicability
    # Generate (x1,x2,g) triples:
    X1 = np.array([np.random.normal(1,noise,N),np.random.normal(1,noise,N)])
    X2 = np.array([np.random.normal(3,noise,N),np.random.normal(2,noise,N)])
    X3 = np.array([np.random.normal(5,noise,N),np.random.normal(3,noise,N)])
    # Concatenate into one data frame
    gdf1 = pd.DataFrame({'x1':X1[0,:],'x2':X1[1,:],'group':'a'})
    gdf2 = pd.DataFrame({'x1':X2[0,:],'x2':X2[1,:],'group':'b'})
    gdf3 = pd.DataFrame({'x1':X3[0,:],'x2':X3[1,:],'group':'c'})
    df = pd.concat([gdf1,gdf2,gdf3],axis=0)
    return df

df0_125 = createData(0.125)
df0_25 = createData(0.25)
df0_5 = createData(0.5)
df1_0 = createData(1.0)
df2_0 = createData(2.0)
````

2. Make scatterplots of the  (ùëã1,ùëã2)  points by group for each of the datasets. As the noise goes up from 0.125 to 2.0, what happens to the visual distinctness of the clusters?

````python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
````

````python
datasets = {
    'df0_125': df0_125,
    'df0_25': df0_25,
    'df0_5': df0_5,
    'df1_0': df1_0,
    'df2_0': df2_0
}

# Scatter plots for each dataset
fig, axes = plt.subplots(1, 5, figsize=(20, 4))
for ax, (name, df) in zip(axes, datasets.items()):
    sns.scatterplot(data=df, x='x1', y='x2', hue='group', ax=ax)
    ax.set_title(name)
plt.tight_layout()
plt.show()

````

3. Create a scree plot for each of the datasets. Describe how the level of noise affects the scree plot (particularly the presence of a clear "elbow") and your ability to definitively select a  ùëò . (Pay attention to the vertical axis across plots, or put all the scree curves on a single canvas.)

````python
from sklearn.cluster import KMeans
fig, ax = plt.subplots(figsize=(8, 5))

for name, df in datasets.items():
    X = df[['x1', 'x2']].values
    sse = []
    k_range = range(1, 10)

    for k in k_range:
        kmeans = KMeans(n_clusters=k, random_state=0, n_init=10)
        kmeans.fit(X)
        sse.append(kmeans.inertia_)

    ax.plot(k_range, sse, marker='o', label=name)

ax.set_xlabel('Number of clusters (k)')
ax.set_ylabel('Sum of Squared Errors (SSE)')
ax.set_title('Scree Plots for Different Noise Levels')
ax.legend()
plt.show()

````

4. Explain the intuition of the elbow, using this numerical simulation as an example.

````python
# Example for one dataset (df0_125) to explain the elbow method
X = df0_125[['x1', 'x2']].values
sse = []
k_range = range(1, 10)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=0, n_init=10)
    kmeans.fit(X)
    sse.append(kmeans.inertia_)

plt.figure(figsize=(6, 4))
plt.plot(k_range, sse, marker='o', linestyle='-', color='b')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Sum of Squared Errors (SSE)')
plt.title('Scree Plot for df0_125 (Example of Elbow Method)')
plt.show()
````



> The elbow method helps determine optimal k by analyzing the SSE. In the scree plot I generated above, SSE decreases rapidly from k = 1 to k = 3, then flattens out. This sharp drop followed by a gradual decline forms an "elbow" at k = 3, indicating that adding more clusters beyond this point does not significantly reduce SSE. This means that k = 3 is the optimal choice, as it balances accuracy and simplicity. Beyond this, additional clusters capture noise rather than meaningful structure.

1.2. Load the data and clean it; maxmin normalize the data and remove any nan's (KMeans from sklearn doesn't accept nan input).

````python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import seaborn as sns
import matplotlib.pyplot as plt

# Load Airbnb dataset
df = pd.read_csv('/content/airbnb_hw.csv')

df = df.dropna(subset=['Price'])  # Remove rows where Price is NaN

#print(df.head())
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')

print(df.head())

# Select relevant features for clustering
df = df[['Price', 'Beds', 'Number Of Reviews', 'Review Scores Rating']]

# Remove missing values (K-Means does not handle NaN values)
df = df.dropna()

# Define Min-Max normalization function
def maxmin(x):
    return (x - x.min()) / (x.max() - x.min())

# Apply Min-Max normalization to the numerical columns
df[['Price', 'Beds', 'Number Of Reviews', 'Review Scores Rating']] = df[['Price', 'Beds', 'Number Of Reviews', 'Review Scores Rating']].apply(maxmin)

# Display the first few rows to check the data
print(df.head())
````

3. Use sklearn's KMeans module to cluster the data by Beds, Number of Reviews, and Review Scores Rating for k=6.

````python
from sklearn.cluster import KMeans

# Select features for clustering
X = df[['Beds', 'Number Of Reviews', 'Review Scores Rating']]

# Apply K-Means clustering with k=6
kmeans = KMeans(n_clusters=6, random_state=42, n_init=10)
df['Cluster'] = kmeans.fit_predict(X)

# Display the mean values of each cluster
print(df.groupby('Cluster').mean())
````

4. Use seaborn's .pairplot() to make a grid of scatterplots that show how the clustering is carried out in multiple dimensions.

````python
sns.pairplot(df, hue='Cluster', diag_kind='kde')
plt.show()
````

5. Use .groupby and .describe to compute the average price for each cluster. Which clusters have the highest rental prices?

````python
# Compute summary statistics for each cluster
cluster_summary = df.groupby('Cluster')['Price'].describe()
print(cluster_summary)

# Identify the cluster with the highest average rental price
highest_price_cluster = cluster_summary['mean'].idxmax()
````

6. Use a scree plot to pick the number of clusters and repeat steps 4 and 5.

````python
sse = []  # List to store sum of squared errors
k_range = range(1, 11)  # Test k values from 1 to 10

# Compute SSE for different k values
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X)
    sse.append(kmeans.inertia_)  # Store SSE (inertia)

# Plot Scree Plot (Elbow Method)
plt.figure(figsize=(6, 4))
plt.plot(k_range, sse, marker='o', linestyle='-', color='b')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Sum of Squared Errors (SSE)')
plt.title('Scree Plot for Optimal k')
plt.show()
````
